4. Call SheCodes Weather API
5. Display weather in the origin and destination

```python
import requests
from rich import print
from rich.markdown import Markdown
from rich.color import Color

def display_current_weather(location):
  # Get the real time current temperature and forecast for the location
  api_key = "885f433f6b0e4583af66o3432taca43c"
  api_url = f"https://api.shecodes.io/weather/v1/current?query={location}&key={api_key}&units=metric"

  response = requests.get(api_url)
  response_data = response.json()

  temperature = round(response_data['temperature']['current'])
  forecast = response_data['condition']['description']

  print(f"The current temperature in [bold]{location}[/bold] is [bold]{temperature}[/bold]ÀöC and {forecast}")



def generate_itinerary(origin, destination, duration, destination_days, travel_time):
  # Generate travel itinerary between origin and destination using AI with tourist attraction stops
  print(f"\n\nGenerating itinerary from {origin} to {destination}... \n")

  context = "You are a helpful travel agent. You know all the best hidden gems to visit. You are famous for your recommended stops on the way to a destination."
  prompt = f"Please create a travel itinerary from {origin} to {destination} in {duration} days with {destination_days} spent in {destination} and {travel_time} spent traveling to {destination}. Keep the itinerary short, less than 20 lines. Use a maximum of one emoji per line and be friendly. Add an estimated price in the country's currency for each day"
  api_key = "885f433f6b0e4583af66o3432taca43c"
  api_url = f"https://api.shecodes.io/ai/v1/generate?prompt={prompt}&context={context}&key={api_key}"

  response = requests.get(api_url)
  response_data = response.json()
  itinerary = Markdown(response_data['answer'])

  print(itinerary)

def welcome():
  # Welcome message to greet user
  print("Welcome to the AI Travel Itinerary Planner")


def credit():
  """ Credit message with developer's name """
  print("[turquoise4] The AI Travel Itinerary Planner was built by [bold]Maryanne McGlone[/bold], bon voyage![/turquoise4]")

welcome()
credit()

# User inputs
origin = input("From where would you like to start your trip? ")
destination = input("What city is your final destination? ")
duration = input("Enter the number of days you will be traveling, (ie 5)... ")
destination_days = input("How many days do you want to spend at your destination? ")

print(f"\n\n{origin}, {destination}, {duration} day trip, {destination_days} days in {destination}\n")

travel_time = int(duration)- int(destination_days)
if travel_time > 0 and origin and destination and duration.isdigit() and destination_days.isdigit():
  display_current_weather(origin)
  display_current_weather(destination)
  generate_itinerary(origin, destination, duration, destination_days, travel_time)
else:
  print("Please try again, make sure you enter valid information")



```
